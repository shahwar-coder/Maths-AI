## 2Ô∏è‚É£ Navigate to Project Folder

**Command:**
```bash
cd ~/Documents/Math\ Folder
```

**What this does:**
- Moves the terminal into the folder where maths notebooks and environment will live
- Ensures the environment is created in the correct location

---

## 3Ô∏è‚É£ Create a Virtual Environment

**Command:**
```bash
python3 -m venv ai-maths
```

**What this does:**
- Creates an isolated Python environment inside `ai-maths/`
- Copies a Python interpreter into `ai-maths/bin/python`
- Keeps packages independent from system Python

---

## 4Ô∏è‚É£ Activate the Virtual Environment

**Command:**
```bash
source ai-maths/bin/activate
```

**What this does:**
- Tells the terminal to use `ai-maths/bin/python`
- Ensures `pip install` installs packages into this environment only

---

## 5Ô∏è‚É£ Install Jupyter Kernel Support

**Command:**
```bash
pip install ipykernel
```

**What this does:**
- Installs `ipykernel` inside the environment
- Allows the interpreter to be used as a Jupyter kernel

---

## 6Ô∏è‚É£ Register the Environment as a Jupyter Kernel

**Command:**
```bash
python -m ipykernel install --user \
  --name ai-maths \
  --display-name "Python (AI Maths)"
```

**What this does:**
- Registers the environment with Jupyter
- Makes it selectable inside VS Code

---

```
Select Kernel ‚Üí Python Environments ‚Üí Python (AI Maths)
```
