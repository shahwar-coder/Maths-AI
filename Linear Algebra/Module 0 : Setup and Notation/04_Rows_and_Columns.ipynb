### Row and Column (their correctness matters)

```python
row = np.array([[1, 2, 3]])      # (1, 3)
col = np.array([[1], [2], [3]]) # (3, 1)

print(f"Result of row @ col: {row @ col}")
print(f"Shape of result: {(row @ col).shape}")  # (1, 1)

print(f"Result of col @ row:\n{col @ row}")
print(f"Shape of result: {(col @ row).shape}")  # (3,

# üìå Same values.
# üìå Completely different math.
# üìå Shape controls behavior.
```

### Broadcasting : Powerful but dangerous

```python
# Broadcasting :
#  
# Broadcasting is a feature in NumPy 
# - that allows arrays with different shapes to be used in arithmetic operations.
# It automatically "stretches" smaller arrays to match the shape of larger ones, enabling element-wise operations.
# However, it can be dangerous because it silently changes behavior without errors.


# Broadcasting Example :
A = np.array([[1, 2, 3, 4, 5, 6]])   # (1, 6) row vector
b = np.array([[10, 20, 30, 40, 50, 60]])  # (1, 6) row vector
C = A + b  # Both have shape (1, 6), so they align perfectly
print("NO BRADCASTING HERE...")
print("C (A + b) :", C)  
print("C (A + b) shape:", C.shape)  # Output: (1, 6)

# ------------------------------------------------------------------------

b = np.array([[10], [20], [30], [40], [50], [60]])  # (6, 1) column vector
D = A + b  # Broadcasting occurs here
print("BROADCASTING HERE...")
print("D (A + b) :", D)
print("D (A + b) shape:", D.shape)  # Output: (6, 6)
```
