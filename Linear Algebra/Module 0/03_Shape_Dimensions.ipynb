```python
'''
Q1. What does ‚Äúdimension‚Äù mean in simple words?
Ans. Dimension tells us how many directions the data has.
'''
# Example
# Scalar ‚Üí 0-D (just one number)
# Vector ‚Üí 1-D (a list)
# Matrix ‚Üí 2-D (rows and columns)


'''
Q2. What is the dimension of a scalar, vector, and matrix?
Ans. Scalar has 0 dimensions, vector has 1 dimension, and matrix has 2 dimensions.
'''
# Example
# 5           ‚Üí scalar ‚Üí 0-D
# [2, 4, 6]   ‚Üí vector ‚Üí 1-D
# [[1,2],[3,4]] ‚Üí matrix ‚Üí 2-D


'''
Q3. What does ‚Äúshape‚Äù mean?
Ans. Shape tells the exact size of the data: how many rows, columns, or elements.
'''
# Example
# Shape answers: ‚ÄúHow big exactly?‚Äù


'''
Q4. What is the shape of a vector like [2, 4, 6]?
Ans. Its shape is (3,) because it has 3 elements.
'''
# Example
vector = [2, 4, 6]
# shape ‚Üí (3,)
# Here the shape is not (1, 3) because it is a 1-D array, not a 2-D array with 1 row.
# So the dimensions matter, (3,) is 1-D, wereas (1, 3) is 2-D.
# ----------
# Row vector (2-D)
row_vector = [[2, 4, 6]]
# shape ‚Üí (1, 3)

# Column vector (2-D)
column_vector = [[2], [4], [6]]
# shape ‚Üí (3, 1)

# Note: IMPORTANT
# Outer brackets = rows
# Inner brackets = columns


'''
Q5. What is the shape of a matrix with 2 rows and 3 columns?
Ans. The shape is written as (2, 3).
'''
# Example
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]
# shape ‚Üí (2, 3)


'''
Q6. Why does shape matter in AI?
Ans. Because matrix operations only work when shapes are compatible.
'''
# Example
# X shape = (100, 784)
# W shape = (784, 10)
# X ¬∑ W ‚Üí (100, 10) ‚úîÔ∏è


'''
Q7. What happens if shapes don‚Äôt match?
Ans. The math fails and the model cannot run.
'''
# Example
# (100, 784) ¬∑ (500, 10) ‚ùå
# Inner numbers don‚Äôt match (784 ‚â† 500)


'''
Q8. What is the simple rule for matrix multiplication?
Ans. The inside numbers must match.
'''
# Example (rule)
# (A, B) ¬∑ (B, C) ‚Üí (A, C)
```

### Scaler vs Vector vs Matrix

```python
import numpy as np

scaler = np.array(5)
vector = np.array([2, 4, 6])
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

print(f"Shape (Scalar): {scaler.shape}, Dimensions (Scalar): {scaler.ndim}")
print(f"Shape (Vector): {vector.shape}, Dimensions (Vector): {vector.ndim}")
print(f"Shape (Matrix): {matrix.shape}, Dimensions (Matrix): {matrix.ndim}")
```

### Row vs Column

```python
row_vector = np.array([[2, 4, 6]])
column_vector = np.array([[2], [4], [6]])

print(row_vector.shape)     # (1, 3)
print(column_vector.shape)  # (3, 1)

# Shape depends on brackets, not on how it looks on paper
```

### Why shape matter in AI ? (Core Idea)

```python
X = np.random.rand(100, 784)
W = np.random.rand(784, 10)

Y = X @ W
print(Y.shape)  # (100, 10)

# üìå Interpretation:
# 100 samples
# 10 output classes
# This is exactly how neural networks work internally.
```

```python
X = np.random.rand(100, 784)
W = np.random.rand(500, 10)

# X @ W   # ‚ùå This will raise an error
# Inner numbers don‚Äôt match (784 ‚â† 500)
```
