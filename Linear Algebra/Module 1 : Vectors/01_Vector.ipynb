### Geometry and Data Connection

### 1. Angle --> Similarity

```python
from numpy.linalg import norm

a = np.array([1, 0])
b = np.array([0.9, 0.1]) # Both these vectors are almost pointing the same way.

cos_sim = (a @ b) / (norm(a) * norm(b)) # cos_sim = (a Â· b) / (|a| * |b|) , # How small is the angle between these two arrows?

print(cos_sim)  # close to 1 â†’ very similar

key_points = '''
Big idea:
- Data can be treated like arrows (vectors).

Angle â†” Meaning:
- Small angle  â†’ arrows point in same direction â†’ similar meaning
- Large angle  â†’ arrows point in different direction â†’ different meaning

Cosine similarity:
- Measures the angle between two vectors
- Value close to 1  â†’ very similar
- Value close to 0  â†’ unrelated

Why important:
- Used in text similarity, search, recommendations, embeddings
- Same math, different meaning (geometry â†’ data)
'''
```

### 2. Distance --> Difference

```python
distance = norm(a - b)
print(distance)

# Large distance â†’ very different items
# Small distance â†’ similar items

key_points = '''
Distance measures difference between vectors.

distance = ||a - b||  â†’ length of the difference vector

Small distance  â†’ vectors are close â†’ similar items
Large distance  â†’ vectors are far  â†’ very different items

In AI:
- vectors = embeddings
- distance = difference in meaning

Rule to remember:
Distance = difference
Closer vectors = more similar
'''
```
