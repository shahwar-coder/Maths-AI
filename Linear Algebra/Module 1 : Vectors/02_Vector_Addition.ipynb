## Vector Addition

### Simple Vector Addition

```python
import numpy as np

v = np.array([2, 1])
u = np.array([1, 3])

print(v + u)   # [3 4]

# üìå Each feature is updated independently.
# AI intuition:
# Combine features
# Adjust embeddings
# Update states
```

### Vector Addition: Shape Matters (Broadcasting Exists)

```python
import numpy as np

np.array([1, 2]) + np.array([3, 4])   # ‚úî works ‚Üí [4 6]
np.array([1, 2]) + np.array([3])      # ‚úî works (broadcasting) ‚Üí [4 5]

# ‚ùå This is NOT allowed (shapes not broadcastable)
np.array([1, 2]) + np.array([3, 4, 5])
# ValueError: operands could not be broadcast together with shapes (2,) (3,)
```
