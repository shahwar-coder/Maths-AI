### Scalar Multiplication

```python
'''
Q1. What is scalar multiplication?
Ans. Scalar multiplication means multiplying a vector by a single number.
'''
# Example
v = [2, 1]
2 * v
# = [4, 2]


'''
Q2. What does scalar multiplication do in geometry?
Ans. It stretches or shrinks the vector without changing its direction.
'''
# Example
# Scalar > 1 â†’ longer vector
# Scalar < 1 â†’ shorter vector
# Scalar < 0 â†’ direction flips


'''
Q3. What does scalar multiplication mean in AI / data terms?
Ans. It controls the importance or strength of a vector.
'''
# Example
# 0.5 * [2, 4]
# = [1, 2]
# Meaning: reduced contribution


q='''
Q4. What is the one-line intuition to remember?
Ans. Addition combines information, scalar multiplication scales importance.
'''
# Example (lock it ðŸ”’)
# Add â†’ combine
# Multiply â†’ scale
```

### Repetition, not Scalar Multiplication (in case of lists)

```python
v = [2, 1]
2 * v
# [2, 1, 2, 1]   âŒ repetition, NOT scalar multiplication
```

### Correct Multiplication (in case of arrays)

```python
import numpy as np

v = np.array([2, 1])
print(2 * v)     # [4 2]
```

### Geometrical Equivalent

```python
v = np.array([2, 1])

print(2 * v)     # longer
print(0.5 * v)   # shorter
print(-1 * v)    # flipped direction
print(0 * v)     # zero vector

# ðŸ“Œ Direction stays same unless scalar is negative.
# ðŸ“Œ Length changes according to scalar magnitude.
# ðŸ“Œ Zero scalar gives zero vector.
```

### AI Intuition

```python
v = np.array([2, 4])

print(0.5 * v)   # [1. 2.]

# ðŸ“Œ In AI this means:
# reduce contribution
# control strength
# adjust influence (learning rate, attention weights, etc.)
```
