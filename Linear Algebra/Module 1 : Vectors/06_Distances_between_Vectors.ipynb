### Question Level 2

```python
'''
Q1. What does distance between vectors mean?
Ans. It means how far apart two vectors (points) are.
Smaller distance  more similar, larger distance  more different.
'''
# Example
# Used to compare words, images, users in AI


'''
Q2. What is Euclidean (L2) distance?
Ans. It is the straight-line distance between two points.
'''
# Formula
# distance(v, u) = sqrt(sum((vi - ui)^2))


'''
Q3. How do you compute Euclidean distance step-by-step?
Ans. Subtract vectors, then take the L2 norm of the result.
'''
# Example
# v = [1, 2], u = [4, 6]
# v - u = [-3, -4]
# distance = sqrt(3 + 4) = 5


'''
Q4. What is Squared Euclidean distance (L2 squared)?
Ans. It is Euclidean distance without the square root.
'''
# Example
# Squared L2 = 3 + 4 = 25
# Faster, keeps same nearest neighbors


'''
Q5. What is Manhattan (L1) distance?
Ans. Distance measured by moving only along axes (grid / city blocks).
'''
# Formula
# L1 = sum(|vi - ui|)
# Example
# |1-4| + |2-6| = 7


'''
Q6. What is Cosine distance?
Ans. It measures how different the directions of two vectors are.
'''
# Formula
# 1 - (A  B) / (||A|| ||B||)
# Ignores vector length


'''
Q7. What does the dot product measure?
Ans. It measures how much two vectors point in the same direction.
'''
# Formula
# A  B = sum(Ai * Bi)
# Larger value  stronger alignment


'''
Q8. When should Euclidean distance be used?
Ans. When both magnitude and direction matter.
'''
# Example
# Image similarity, physical measurements


'''
Q9. When should Cosine distance be used?
Ans. When direction matters but magnitude does not.
'''
# Example
# Word embeddings, text similarity


'''
Q10. What is the most important takeaway about distance metrics?
Ans. Different distance metrics define similarity differently.
'''
# Memory lock 
# L2  straight line
# L1  city blocks
# Cosine  angle
# Dot product  alignment
```

### Vector Distance

```python
# Vector Distance:

'''
- Geometry: distance between two points in space.
- AI: measures how different two items (words, images, users) are.
'''
```

### Vector Distance (Formula)

```python
import numpy as np
from numpy.linalg import norm

v = np.array([3, 4])
u = np.array([0, 0])

distance = norm(v - u)
distance  # Output: 5.0

# This is the Euclidean (L2) distance between vectors v and u.
# This is different from the norm of a single vector, which measures its length.
# This is also different from cosine distance, which measures the angle between vectors.
# This is also different from the dot product, which measures alignment.
# This is also different from Manhattan (L1) distance, which measures distance along axes.
# This is also different from Squared Euclidean distance, which is the square of L2 distance.
```

### AI Intuition

```python
# Small distance  similar items
# Large distance  different items
# Used in k-NN, clustering, recommendations, anomaly detection
```

### Norm vs Distance

```python
# v  size of one vector
# v  u  distance between two vectors
# The distinction is important, because norm measures length, while distance measures difference.
# To imagine the difference, 
# - think of two points on a graph. 
# - The distance between them is the straight-line distance connecting them, 
# - while the norm of one point is just its distance from the origin.
```

### Norm and Distance

```python
# Distance between vectors = norm of their difference
```

### Squared Euclidean distance (L2 squared)

```python
squared_l2 = np.sum((v - u) ** 2)
print(squared_l2) # Output: 25
# This is the Squared Euclidean (L2 squared) distance between vectors v and u
```

### Manhattan (L1) distance

```python
l1_distance = np.sum(np.abs(v - u))
print(l1_distance) # Output: 7
# This is the Manhattan (L1) distance between vectors v and u
```

### Cosine Distance

```python
'''
Cosine distance, Measures difference in direction (ignores magnitude).
'''
a = np.array([1, 1])
b = np.array([2, 2])

cosine_similarity = (a @ b) / (norm(a) * norm(b))
cosine_distance = 1 - cosine_similarity

print(cosine_distance)

# Output: 0.0
# Since a and b point in the same direction, their cosine distance is 0.

# Cosine Similarity and Cosine Distance
# Cosine similarity measures how similar two vectors are in direction.
# Cosine distance measures how different two vectors are in direction.
# Cosine similarity ranges from -1 (opposite) to 1 (same direction).
# Cosine distance ranges from 0 (same direction) to 2 (opposite direction).
```

### Dot Product

```python
dot_product = a @ b
print(dot_product) # Output: 4
# This is the dot product between vectors a and b
# Dot product measures alignment between vectors, meaning how much they point in the same direction.
```
