### Linear Combinations

```python
'''
Q1. What is a linear combination in simple words?
Ans. A linear combination means multiplying vectors by numbers and then adding them.
'''
# Example
# aÂ·v + bÂ·u
# Numbers â†’ scalars
# v, u   â†’ vectors


'''
Q2. How do you write a linear combination mathematically?
Ans. Using scalars multiplied with vectors and added together.
'''
# Example
v = [1, 0]
u = [0, 1]
# 3v + 2u = [3, 2]


'''
Q3. What does a linear combination mean in geometry?
Ans. It means reaching new points by moving along given directions.
'''
# Example
# v = right direction
# u = up direction
# 3v + 2u â†’ move right 3, up 2


'''
Q4. What does a linear combination mean in AI / data view?
Ans. It means mixing features using weights.
'''
# Example
v = [2, 4]
u = [6, 0]
# 0.5v + 0.5u = [4, 2]
# This is a weighted average of features


'''
Q5. Where do linear combinations appear in AI?
Ans. Almost everywhere: neural networks, attention, embeddings, recommendations.
'''
# Example
# output = w1*x1 + w2*x2 + bias
# This is a linear combination


'''
Q6. What is the most important rule for linear combinations?
Ans. All vectors must have compatible shapes.

In pure math: vectors must have the same shape.
In NumPy: broadcasting may allow mixing, but should be used carefully.
'''
# Example
# v = [1, 2]      # shape (2,)
# u = [[3], [4]]  # shape (2,1)
# 2v + 3u â†’ Error in pure math, works in NumPy due to broadcasting


'''
Q7. Why are linear combinations so important for ML?
Ans. Because every neural network layer is a linear combination plus a non-linearity.
'''
# Example
# Linear layer â†’ Wx
# Activation   â†’ f(Wx)


q='''
Q8. What is the one-line idea to remember?
Ans. Linear combinations are weighted mixing of vectors.
'''
# Example (lock it ðŸ”’)
# Multiply â†’ weight
# Add      â†’ mix
```
