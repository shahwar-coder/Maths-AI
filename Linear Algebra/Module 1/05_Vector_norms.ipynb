### Vector Norms

```python
'''
Q1. What is a vector norm in simple words?
Ans. A vector norm is the length or size of a vector.
In AI, it usually means the L2 norm.
'''
# Example
# Norm = how long the arrow is


'''
Q2. What is the L2 norm of a 2D vector?
Ans. It is found using the Pythagorean formula.
'''
# Example
v = [3, 4]
# v = sqrt(3 + 4) = 5


'''
Q3. How do you compute the L2 norm in higher dimensions?
Ans. Square all elements, add them, then take the square root.
'''
# Example
# v = [v1, v2, v3]
# v = sqrt(v1 + v2 + v3)


'''
Q4. What does vector norm mean in AI or data view?
Ans. It represents the strength or magnitude of a signal or embedding.
'''
# Example
# Large norm  strong signal
# Small norm  weak signal


'''
Q5. Where are vector norms used in AI?
Ans. In embeddings, regularization, distance calculation, and normalization.
'''
# Example
# Weight decay uses norm
# k-NN distance uses norm


'''
Q6. Can a vector norm ever be negative?
Ans. No. Vector norms are always greater than or equal to zero.
'''
# Example
# v  0 always


'''
Q7. When is the norm of a vector equal to zero?
Ans. Only when the vector is the zero vector.
'''
# Example
v = [0, 0, 0]
# v = 0


'''
Q8. How does scaling affect the norm?
Ans. Scaling a vector scales its norm by the absolute value of the scalar.
'''
# Example
# 2v = 2  v
# -3v = 3  v


'''
Q9. Why does cosine similarity ignore the norm?
Ans. Because it compares direction (angle) and not magnitude.
'''
# Example
# Same direction, different lengths  cosine similarity = same


q='''
Q10. What is the one-line idea to remember?
Ans. The L2 norm tells how long or strong a vector is.
'''
# Example (lock it )
# Norm = length = strength
```

### L2 Normal in 2D

```python
import numpy as np
from numpy.linalg import norm

v = np.array([3, 4])
l2_norm = norm(v)
print(f"L2 Norm of v: {l2_norm}")  # Output: L2 Norm of v: 5.0

# L2 norm of a vector means its length or magnitude in space.
# Same as Pythagoras.
```

### L2 Normal for 3D, 4D, 5D....(Higher Dimensions)

```python
v = np.array([1, 2, 2])
print(norm(v))   # sqrt(1 + 2 + 2) = 3

#  Works in any dimension.
```
