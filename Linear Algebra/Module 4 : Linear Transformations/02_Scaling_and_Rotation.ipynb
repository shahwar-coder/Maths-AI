### Scaling & Rotation

```python
'''
SETâ€“1:

Q1. What is scaling in terms of vectors and space?
Ans. Scaling means stretching or shrinking space so that vectors become longer or shorter without changing their direction.
'''
# Example
# Scaling up makes vectors move farther from the origin
# Scaling down brings vectors closer to the origin


'''
Q2. What happens to the origin during scaling?
Ans. The origin (0,0) never moves during scaling; it always stays fixed.
'''
# Example
# Any scaling applied to [0, 0] still gives [0, 0]


'''
Q3. How can scaling affect x and y directions differently?
Ans. Scaling can stretch space more in one direction than the other, making shapes wider or taller.
'''
# Example
# Scaling x more â†’ wider shape
# Scaling y more â†’ taller shape


'''
Q4. What is rotation in simple intuitive terms?
Ans. Rotation means turning all vectors around the origin without changing their length or shape.
'''
# Example
# Rotating a vector is like rotating a pinned sheet of paper


'''
Q5. What stays the same and what changes during rotation?
Ans. Length and shape stay the same, but direction changes.
'''
# Example
# A vector keeps its norm but points in a new direction


'''
Q6. How do scaling and rotation differ fundamentally?
Ans. Scaling changes the size of vectors, while rotation changes only their direction.
'''
# Example
# Scaling â†’ zoom in / zoom out
# Rotation â†’ spin effect


q='''
Q7. Why are scaling and rotation important in AI?
Ans. In AI, scaling adjusts feature importance, while rotation changes representation without losing meaning.
'''
# Example
# Weights scale features
# Rotations re-orient embedding spaces
```

```python
'''
SETâ€“2:

Q1. How does a matrix perform scaling?
Ans. A scaling matrix multiplies vector components to increase or decrease their magnitude along specific axes.
'''
# Example
# Scaling matrix:
# [ 2  0 ]
# [ 0  1 ]
# Scales x-direction by 2, y-direction unchanged


'''
Q2. What is uniform scaling?
Ans. Uniform scaling scales all directions by the same factor, preserving the shape of objects.
'''
# Example
# Scaling both x and y by 3 enlarges the shape without distortion


'''
Q3. What is non-uniform scaling?
Ans. Non-uniform scaling uses different scaling factors along different axes, changing the shape.
'''
# Example
# x scaled by 2, y scaled by 1 â†’ object becomes wider


'''
Q4. How does a rotation matrix affect vectors?
Ans. A rotation matrix rotates all vectors by the same angle around the origin while preserving their length.
'''
# Example
# Rotating a vector by 90Â° turns its direction but keeps its magnitude same


'''
Q5. Why does rotation preserve vector norms?
Ans. Because rotation only changes direction, not distance from the origin.
'''
# Example
# A vector of length 5 remains length 5 after rotation


q='''
Q6. Why are scaling and rotation important in machine learning?
Ans. They allow feature spaces to be stretched, compressed, or re-oriented to learn better representations.
'''
# Example
# Neural network layers scale and rotate feature space before applying non-linearity
```

### Let's Scale a vector

```python
import numpy as np

v = np.array([[2],
              [1]])

scale = 3
scaled_v = scale * v
print(scaled_v)

# ðŸ“Œ Bigger scalar â†’ farther from origin
# ðŸ“Œ Smaller scalar â†’ closer to origin
```

### Let's Rotate a vector

```python
theta = np.pi / 2  # 90 degrees

R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

v = np.array([[1],
              [0]])

rotated_v = R @ v
print(rotated_v)

how_we_rotated = """
STEP-BY-STEP: HOW THE VECTOR WAS ROTATED (90Â° ABOUT ORIGIN)

Step 1: Choose the rotation angle
- theta = Ï€/2 radians â†’ 90 degrees
- Positive angle means counter-clockwise rotation

Step 2: Build the rotation matrix
R = [[ cos(theta), -sin(theta) ],
     [ sin(theta),  cos(theta) ]]

- This matrix is specially designed to:
  â€¢ change direction
  â€¢ keep length the same
- It rotates everything around the ORIGIN (0, 0)

Step 3: Define the original vector
v = [1, 0]

- This vector points along the positive x-axis
- Length = 1
- Direction = right

Step 4: Multiply matrix with vector
rotated_v = R @ v

What multiplication means:
- Take x = 1, y = 0
- Compute:
  new_x = cos(theta)*1 + (-sin(theta))*0 = cos(90Â°) = 0
  new_y = sin(theta)*1 + cos(theta)*0  = sin(90Â°) = 1

Step 5: Get the rotated vector
rotated_v = [0, 1]

- Direction changed: right â†’ up
- Length stayed the same: still 1
- Origin stayed fixed

FINAL INTUITION:
- Rotation = turning the arrow
- No stretching
- No shrinking
- Only direction changes
"""

# Output : [[6.123234e-17]
#  [1.000000e+00]]

# This is essentially equivalent to -> [0, 1], #hence rotated.
```

### Rotation does not distort Shape or change Length

```python
import numpy as np

# Original vector
v = np.array([3, 4])   # length = 5

# Rotation angle (90 degrees)
theta = np.pi / 2

# Rotation matrix
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# Rotate the vector
v_rot = R @ v

# Lengths before and after rotation
orig_len = np.linalg.norm(v)
rot_len  = np.linalg.norm(v_rot)

print("Original vector:", v)
print("Rotated vector :", v_rot)
print("Original length:", orig_len)
print("Rotated length :", rot_len)
```

```python
# We can the same thing in different ways
# Rotation preserves shape and length
# Rotation changes direction, not shape or length
# Rotation is a length-preserving (distance-preserving) transformation
```
