### Representation of systems using matrices

```python
'''
Q1. Why do we represent multiple equations using a matrix equation Ax = b?
A.
Because a matrix equation bundles many separate equations into one compact form.
It makes the system easier to analyze, manipulate, and solve.
'''
# Example:
# Equations:
# 2x + y = 5
# x + 3y = 6
# Matrix form:
# A = [[2, 1],
#      [1, 3]]
# x = [x, y]^T
# b = [5, 6]^T
# Ax = b


'''
Q2. What does the matrix A represent in Ax = b?
A.
Matrix A stores all the coefficients of the variables from the system of equations.
Each row corresponds to one equation.
'''
# Example:
# Equation: 2x + y = 5
# Coefficients  [2, 1]
# These coefficients become one row of matrix A


'''
Q3. What does the vector x represent in Ax = b?
A.
Vector x contains the unknown variables we are trying to find.
Each entry corresponds to one variable.
'''
# Example:
# Variables: x and y
# x-vector = [x,
#             y]


'''
Q4. What does the vector b represent in Ax = b?
A.
Vector b contains the constants or results on the right-hand side of each equation.
'''
# Example:
# Equations:
# 2x + y = 5
# x + 3y = 6
# b = [5,
#      6]


'''
Q5. How does each row of Ax = b relate to an equation?
A.
Each row of matrix A multiplied by vector x produces one equation.
All rows together represent the full system.
'''
# Example:
# Row 1: [2, 1]  [x, y] = 5
# Row 2: [1, 3]  [x, y] = 6
# Together they form Ax = b


q='''
Q6. What is an augmented matrix and why is it useful?
A.
An augmented matrix combines the coefficient matrix and result vector into one table.
It is useful for solving systems using row operations.
'''
# Example:
# Augmented matrix:
# [ 2  1 | 5 ]
# [ 1  3 | 6 ]
# Left side  coefficients
# Right side  results
```

```python
'''
Q1. How does Ax = b appear in machine learning problems?
A.
In machine learning, Ax = b represents predicting outputs from data.
A is the data (features), x is the model parameters, and b is the target output.
'''
# Example:
# A = [[hours_studied, classes_attended],
#      [...]]
# x = [weight1, weight2]
# b = [exam_score]
# Ax  b  prediction model


'''
Q2. Why do we write Ax  b instead of Ax = b in ML?
A.
Because real-world data is noisy and imperfect.
There is usually no exact solution, so we look for the closest possible match.
'''
# Example:
# Data points don't lie perfectly on a line
# Model tries to minimize error:
# Ax  b (best fit), not exact equality


'''
Q3. What does solving Ax = b mean in linear regression?
A.
It means finding the weights that best map input features to output values.
This is the core goal of linear regression.
'''
# Example:
# A = data matrix (N samples  D features)
# x = weight vector
# b = labels
# Training = find x such that Ax  b


'''
Q4. How does Ax = b relate to loss functions?
A.
The difference between Ax and b defines the error.
Loss functions measure how far Ax is from b.
'''
# Example:
# Prediction = Ax
# Error = Ax - b
# Loss = ||Ax - b|| (mean squared error)


'''
Q5. Why is Ax = b important for optimization and gradients?
A.
Because optimization adjusts x to reduce the difference between Ax and b.
Gradients tell us how to change x to improve predictions.
'''
# Example:
# Start with random x
# Compute error = Ax - b
# Update x using gradients to reduce error


q='''
Q6. How does Ax = b appear in modern AI systems like RAG or embeddings?
A.
Similarity, scoring, and transformations are all matrix-vector operations.
Ax = b is the basic building block under the hood.
'''
# Example:
# A = embedding matrix (documents)
# x = query embedding
# b = similarity scores
# Ax gives relevance scores for retrieval
```
