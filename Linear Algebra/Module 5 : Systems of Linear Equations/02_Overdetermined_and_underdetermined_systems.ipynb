### Overdetermined vs underdetermined systems

```python
'''SET-1
Q1. What determines whether a system Ax = b is overdetermined or underdetermined?
A.
It depends on the comparison between:
 number of equations (rows of A)
 number of unknowns (size of x)
More equations than unknowns  overdetermined
More unknowns than equations  underdetermined
'''
# Example:
# A has shape (5, 2)
# 5 equations, 2 unknowns
#  overdetermined system


'''
Q2. What is an overdetermined system in simple words?
A.
An overdetermined system has more equations than unknowns.
There are too many constraints trying to decide too few variables.
'''
# Example:
# Equations:  x = 1
#             x = 2
# One variable, two conflicting rules
#  no exact solution


'''
Q3. Why do overdetermined systems usually have no exact solution?
A.
Because different equations often conflict slightly,
so no single value of x can satisfy all equations at the same time.
'''
# Example:
# x + y = 2
# x + y = 2.1
# Both cannot be true simultaneously
#  no exact solution


'''
Q4. What is an underdetermined system in simple words?
A.
An underdetermined system has fewer equations than unknowns.
There are not enough rules to fix a single answer.
'''
# Example:
# x + y = 10
# Two unknowns, one equation
#  infinitely many solutions


'''
Q5. Why do underdetermined systems have infinitely many solutions?
A.
Because extra variables are free to change
without violating the given equations.
'''
# Example:
# x + y = 10
# x = 1  y = 9
# x = 2  y = 8
# Many valid solutions exist


q='''
Q6. What is a square (just-right) system and why is it special?
A.
A square system has the same number of equations and unknowns.
It may have:
 one solution
 no solution
 infinitely many solutions
depending on the equations.
'''
# Example:
# x + y = 4
# x - y = 2
# Two equations, two unknowns
#  exactly one solution (x=3, y=1)
```

```python
'''SET-2
Q1. Why are most machine learning problems overdetermined?
A.
Because ML usually has many data points (equations)
but only a limited number of model parameters (unknowns).
So there are more equations than variables.
'''
# Example:
# 10,000 training samples  10,000 equations
# Model has 100 parameters  100 unknowns
# Ax = b becomes overdetermined
#  exact solution unlikely


'''
Q2. Why does ML solve Ax  b instead of Ax = b?
A.
Because in overdetermined systems, equations conflict due to noise.
So ML finds the best approximate solution, not an exact one.
'''
# Example:
# Data points dont lie perfectly on a line
# Linear regression minimizes error instead of solving exactly


'''
Q3. How does least squares relate to overdetermined systems?
A.
Least squares finds the x that minimizes the total error
between Ax and b when no exact solution exists.
'''
# Example:
# Minimize ||Ax  b||
# Instead of forcing Ax = b
#  best compromise solution


'''
Q4. Why are deep learning models often underdetermined?
A.
Because they have more parameters than training examples,
creating many possible solutions that fit the data.
'''
# Example:
# 1,000 training samples
# 10 million neural network parameters
#  underdetermined system


'''
Q5. Why is regularization needed in underdetermined ML systems?
A.
Because with many solutions available,
regularization helps choose a preferred one
(e.g., smallest weights or simplest model).
'''
# Example:
# L2 regularization chooses the minimum-norm solution
# among infinitely many valid solutions


q='''
Q6. How do overdetermined and underdetermined systems appear in RAG or embeddings?
A.
Similarity search is often overdetermined (many documents, one query),
while embedding spaces are underdetermined (many dimensions, few constraints).
'''
# Example:
# RAG:
# Query vector compared against thousands of document vectors  overdetermined
# Embedding training:
# Few labels but high-dimensional vectors  underdetermined
```
