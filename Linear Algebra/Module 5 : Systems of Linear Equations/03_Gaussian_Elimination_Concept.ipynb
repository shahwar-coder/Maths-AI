### Gaussian Elimination

```python
'''
Q1. What is Gaussian elimination in simple words?
A.
Gaussian elimination is a step-by-step method to simplify a system of equations
until the solution becomes easy to see.
It removes confusion by reducing equations one variable at a time.
'''
# Example:
# Start with:
# 2x + y = 5
# x + 3y = 6
# Rearrange and simplify step by step to make solving easier


'''
Q2. What problem does Gaussian elimination try to solve?
A.
It solves the problem of many equations with many unknowns that are tangled together.
The goal is to untangle them into a simpler form.
'''
# Example:
# Ax = b
# Many equations → simplify → clear path to x


'''
Q3. What does “elimination” actually mean?
A.
Elimination means using one equation to remove a variable from another equation,
so each equation becomes simpler than before.
'''
# Example:
# Eq1: x + y = 4
# Eq2: x - y = 2
# Use Eq1 to eliminate x from Eq2


'''
Q4. Why does Gaussian elimination aim for a staircase shape?
A.
The staircase (step-like) form ensures each equation focuses on a new variable.
This makes solving variables one-by-one straightforward.
'''
# Example:
# First row → solve x
# Second row → solve y
# Third row → solve z


'''
Q5. Why are row operations allowed in Gaussian elimination?
A.
Each row represents an equation.
Changing rows means rewriting equations in an equivalent way without changing the solution.
'''
# Example:
# Swapping equations or adding equations together
# does NOT change the final solution


q='''
Q6. Why is Gaussian elimination important conceptually for ML?
A.
Even though ML rarely solves equations exactly, Gaussian elimination teaches
how to simplify complex problems into smaller, manageable steps.
'''
# Example:
# Exact solving → Gaussian elimination
# Approximate solving → optimization & least squares
```

### Gaussian Elimination in Action

```python
# 2x + y = 5
#  x + 3y = 6

# Step 1: Eliminate x
# → multiply second equation by -2 and add

# (2x + y) + (-2x - 6y) = 5 - 12
# → -5y = -7
# → y = 7/5

# Step 2: Plug back
# → x = (5 - y)/2
```
